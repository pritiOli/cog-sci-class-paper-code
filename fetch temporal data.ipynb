```python
import os
country_of_interest=['ASM','UMI','GUM','MNP','PRI','USA','GBR','FRA','GUF','PYF','ATF','GLP','NCL','MTQ','MYT','DEU','NZL',
                    'NIU','COK','TKL','RUS','MAC','HKG','CHN','IND','ISR','ZAF']
```

```python
path ='../all_data/Baturo for topic modeling'
```

```python
all_files=os.listdir(path)
all_years=set()
```

```python
for file in all_files:
    year=file.split('.txt')[0][-4:]   
    all_years.add(year)
```

```python
print(all_years)
```

```python
for year in all_years:
    open('temporal data interest/'+year+'.txt','w+',encoding='utf-8')
```

```python
for file in all_files:
    country=file[:3]
    year=file.split('.txt')[0][-4:] 
    try:
        content= open(path+'/'+file,'r', encoding='utf-8').read()
        write_to = open('temporal data/'+year+'.txt','a',encoding='UTF-8')
        write_to_interest = open('temporal data interest/'+year+'.txt','a',encoding='UTF-8')
        write_to.write(content)
        if country in country_of_interest:
            write_to_interest.write(content)
        print(f"successfully added the content of {file}.txt to {year}.txt")
    except:
        print(f'error occured for file {file}')
```
