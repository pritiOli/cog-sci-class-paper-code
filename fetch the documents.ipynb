```python
import os
from shutil import copyfile

all_files=[]
country_of_interest=['ASM','UMI','GUM','MNP','PRI','USA','GBR','FRA','GUF','PYF','ATF','GLP','NCL','MTQ','MYT','DEU','NZL',
                    'NIU','COK','TKL','RUS','MAC','HKG','CHN','IND','ISR','ZAF']
BASE_DIR='.\Baturo UNGA'
folders = os.listdir(BASE_DIR)
session_folders = [ folder for folder in folders if folder[:7]=='Session' ]
```

```python
for folder in session_folders:
    file_dir=os.path.join(BASE_DIR,folder)
    files_dir = os.listdir(file_dir)
    for file in files_dir:
        if file[:3] in  country_of_interest:
            all_files.append(os.path.join(file_dir,file))
```
